# TP3 : Étude de la concentration des distances en haute dimension (Corrigé)

Ce corrigé présente la simulation et l'analyse de la concentration des distances dans des espaces de grande dimension.

## Objectifs et démarche
- Simuler des points dans des espaces de différentes dimensions
- Calculer les distances moyennes et écarts-types
- Visualiser et interpréter la concentration des distances

```python
# 1. Simuler des points dans [0,1]^d pour plusieurs valeurs de d
import numpy as np
import matplotlib.pyplot as plt
dims = [2, 5, 10, 50, 100]
n_points = 1000
results = {}
for d in dims:
    X = np.random.rand(n_points, d)
    D = np.sqrt(((X[None,:,:] - X[:,None,:])**2).sum(axis=2))
    D_flat = D[np.triu_indices(n_points, k=1)]
    mean = D_flat.mean()
    std = D_flat.std()
    results[d] = (mean, std)
```

```python
# 2. Visualiser l'évolution du rapport écart-type/moyenne
ratios = [results[d][1]/results[d][0] for d in dims]
plt.figure(figsize=(8,6))
plt.plot(dims, ratios, marker='o')
plt.xlabel('Dimension d')
plt.ylabel('Écart-type / Moyenne')
plt.title('Concentration des distances en fonction de la dimension')
plt.grid(True)
plt.show()
```

## Analyse et interprétation
- Lorsque la dimension augmente, le rapport écart-type/moyenne diminue fortement.
- Les distances deviennent de plus en plus similaires : c'est la concentration des distances.
- Ce phénomène illustre la malédiction de la dimension.
